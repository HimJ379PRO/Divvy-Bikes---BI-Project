## Let us use *pandas* to combine quarterly datasets for the years 2018 and 2019 each into a single dataset.
We will combine the total of 8 datasets containing millions of records individually into a single dataset containing upto 2 million records, so that the Tableau Public dashboard works at relatively faster speed. It will also take care of 15 million records limit of Tableau public. 

Data downloaded from [Divvy Official website]( https://divvybikes.com/system-data)

```python
# Import DSc libraries.
import pandas as pd
```

```python
# Read the 2018 Q1 Trips table.
df_2018Q1 = pd.read_csv("C:\\Users\Himanshu\Downloads\Divvy 2018\Divvy_Trips_2018_Q1.csv")

# Check the table.
df_2018Q1
```

```python
#Check for NULLS in the important columns.
df_2018Q1.info()
```

### No NULLS in the important columns. We can ignore the NULLS in the columns containing values for 'Gender' and 'Birthyear'.

Proceed to *sample* the data

```python
# Sample the data in the 2018 Q1 dataset.
df_sampled_2018Q1 = df_2018Q1.sample(frac=0.25, random_state=3)
# Check the sampled dataset.
df_sampled_2018Q1
```

```python
# Check the dataset info and verify data types of all the columns. Also, check the total records.
df_sampled_2018Q1.info()
```

### Everything looks correct*. We **did not check** for the **'accuracy' and 'consistency' of values in the individual fields.** The same would be verified before and at the time of loading the data into a postgreSQL database using **constraints**.  
### Now, we will repeat the above steps for the data on remaining quarters of 2018.

```python
df_2018Q2 = pd.read_csv("C:\\Users\Himanshu\Downloads\Divvy 2018\Divvy_Trips_2018_Q2.csv")
df_2018Q3 = pd.read_csv("C:\\Users\Himanshu\Downloads\Divvy 2018\Divvy_Trips_2018_Q3.csv")
df_2018Q4 = pd.read_csv("C:\\Users\Himanshu\Downloads\Divvy 2018\Divvy_Trips_2018_Q4.csv")
```

```python
df_2018Q2.info()
df_2018Q3.info()
df_2018Q4.info()
```

```python
df_sampled_2018Q2 = df_2018Q2.sample(frac=0.25, random_state=3)

df_sampled_2018Q2
```

```python
# Combine the 2 sampled datasets to check whether we are getting the intended result.
test_df_2018 = pd.concat([df_sampled_2018Q1, df_sampled_2018Q2],ignore_index= True)
# Check the combined dataset for 2018 Q1 and Q2
test_df_2018
```

```python
# Check the dataset info.
test_df_2018.info()
```

### We got into a problem!!! The columns in the datasets are named differently. *pandas* combined the datasets horizontly. 
### To combine the data vertically, we will rename the columns across all the datasets consistently whenever needed.

```python
# Rename the columns in 2018 Q1 to match across all the datasets
df_sampled_2018Q1.rename(columns={'01 - Rental Details Rental ID': 'trip_id',
                                  '01 - Rental Details Local Start Time': 'start_time',
                                  '01 - Rental Details Local End Time': 'end_time',
                                  '01 - Rental Details Bike ID': 'bikeid',
                                  '01 - Rental Details Duration In Seconds Uncapped': 'tripduration',
                                  '03 - Rental Start Station ID': 'from_station_id',
                                  '03 - Rental Start Station Name': 'from_station_name',
                                  '02 - Rental End Station ID': 'to_station_id',
                                  '02 - Rental End Station Name': 'to_station_name',
                                  'User Type': 'usertype',
                                  'Member Gender': 'gender',
                                  '05 - Member Details Member Birthday Year': 'birthyear'}, inplace=True)
# Check the renamed columns
df_sampled_2018Q1.info()
```

```python
# Recombine the datasets to create a test dataset.
test_df_2018 = pd.concat([df_sampled_2018Q1, df_sampled_2018Q2],ignore_index= True)
# Check the dataset info.
test_df_2018.info()
```

### AWESOME!!! We solved the problem. Now, we will repeat the steps for the remaining datasets.

```python
df_sampled_2018Q3 = df_2018Q3.sample(frac= 0.25, random_state= 3)
df_sampled_2018Q4 = df_2018Q4.sample(frac= 0.25, random_state= 3)

df_sampled_2018Q3.info()
df_sampled_2018Q4.info()
```

### Great! No NULLS in the important columns. We can ignore the NULLS in the columns containing values for 'Gender' and 'Birthyear'.
### We will now create a combined dataset for the year 2018.

```python
# Combine all 2018 Q1,Q2,Q3,Q4 datasets into a single dataset.
df_2018 = pd.concat([df_sampled_2018Q1, df_sampled_2018Q2, df_sampled_2018Q3, df_sampled_2018Q4], ignore_index= True)
# Check the table.
df_2018
```

### PERFECT! Now, we will repeat all the above steps to create a dataset for the year 2019.

```python
# Read the datasets
df_2019Q1 = pd.read_csv("H:\\My Drive\Business\Datasets\Divvy Bikes\Divvy_Trips_2019_Q1.csv")
df_2019Q2 = pd.read_csv("H:\\My Drive\Business\Datasets\Divvy Bikes\Divvy_Trips_2019_Q2.csv")
df_2019Q3 = pd.read_csv("H:\\My Drive\Business\Datasets\Divvy Bikes\Divvy_Trips_2019_Q3.csv")
df_2019Q4 = pd.read_csv("H:\\My Drive\Business\Datasets\Divvy Bikes\Divvy_Trips_2019_Q4.csv")
```

```python
# Check the datasets
df_2019Q1.info()
df_2019Q2.info()
df_2019Q3.info()
df_2019Q4.info()
```

```python
# Sample the datasets
df_sampled_2019Q1 = df_2019Q1.sample(frac = 0.25, random_state=3)
df_sampled_2019Q2 = df_2019Q2.sample(frac = 0.25, random_state=3)
df_sampled_2019Q3 = df_2019Q3.sample(frac = 0.25, random_state=3)
df_sampled_2019Q4 = df_2019Q4.sample(frac = 0.25, random_state=3)
```

```python
# Combine all 4 sampled datasets into 1 single sampled dataset for year 2019
df_2019 = pd.concat([df_sampled_2019Q1, df_sampled_2019Q2, df_sampled_2019Q3, df_sampled_2019Q4],ignore_index= True)
# Check the table.
df_2019
```

### Let's COMBINE the datasets for the years 2018 and 2019 to create a single dataset.

```python
# Combine both the sampled 2018 and 2019 datasets into 1 single dataset
df_2018_2019 = pd.concat([df_2018, df_2019], ignore_index= True)
# Check the dataset info.
df_2018_2019.info()
```

```python
# Check the table.
df_2018_2019
```

### Save the final dataset into a CSV file.

```python
# Export the dataset to a CSV file
df_2018_2019.to_csv('Divvy_Trips.csv',index=False)
```
